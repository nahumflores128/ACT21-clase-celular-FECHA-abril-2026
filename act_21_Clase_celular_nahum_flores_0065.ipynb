```python
class Celular:
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} se ha encendido.")

    def mostrar_info(self):
        print(f"Celular: {self.marca} {self.modelo}, Precio: ${self.precio}")

# Creacion de objetos (Instancias)
cel1 = Celular("Samsung", "Galaxy S23", 800)
celular2 = Celular("Apple", "iphone 15", 1000)
# llamadas a las funciones para cada objeto celular 1 y celular 2
print("Celular 1:")
cel1.mostrar_info()
cel1.encender()

print("\nCelular 2:")
celular2.mostrar_info()
celular2.encender()

print("progama terminado")
print("Creador de software por Nahum flores NL = 18")
```
